Step 1. Import the necessary libraries

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
```

Step 3. Assign it to a variable called crime.

```python
#Reading a csv file
url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime=pd.read_csv(url)
crime.head()
```

Step 4. What is the type of the columns?

```python
# What is the type of columns
crime.info()
```

Step 5. Convert the type of the column Year to datetime64

```python
# converting the type of colum year to datetime64
crime.Year=pd.to_datetime(crime.Year, format='%Y')
crime.head()
```

```python
crime.info()
```

Step 6. Set the Year column as the index of the dataframe

```python
# set the year columns as the index of the dataframe
crime=crime.set_index('Year',drop=True)

crime.head()
```

Step 7. Delete the Total column

```python
# deleting the column
del crime["Total"]
crime.head()
```

Step 8. Group the year by decades and sum the values

```python
# gropny-group the year by decades and sum the values
crime=crime.resample('10AS').sum()
population=crime['Population'].resample('10AS').max()
crime['Population']=population
crime
```
